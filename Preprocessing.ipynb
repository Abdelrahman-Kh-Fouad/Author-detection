```python
import nltk
```

```python
firstBookFile = open("./books/Book 1 - The Philosopher's Stone.txt")
firstFileRaw = firstBookFile.read()
firstFileRaw
```

```python
#Delete Page Numbers
import re 
firstFileRaw = re.sub(r'\bPage \| .*\b','',firstFileRaw)
print(firstFileRaw)
```

```python
#Delete multilines
firstFileRaw = re.sub(r'\n[\n]*','\n',firstFileRaw)
print(firstFileRaw)
```

```python
#Delete chapters' names 
chapter =1 
fileReadyToTok = ""
for line in firstFileRaw.split('\n'):
    line = line.strip()
    if(line ==''):
        continue
    allUpper =True
    for flag in [(lambda ch : ch.isspace() or(ch.isalpha() and ch.isupper()))(ch) for ch in line ]:
        allUpper = allUpper and flag
    if(allUpper):
        print(f'chapter {chapter}-> {line}')
        chapter+=1
    else :
        fileReadyToTok+=line+" "
```

```python
#Lowering all chars 
tmp = ''
for ch in fileReadyToTok:
    to=ch
    if(ch.isalpha()):
        to = ch.lower()
    tmp+=to
fileReadyToTok = tmp 
print(fileReadyToTok)
```

```python
words = fileReadyToTok.split(' ')
print(len(words))
print(words)
```

```python
#Filtering from stopwords
from nltk.corpus import stopwords
stopwordsList = stopwords.words('english')
newWords = []
for word in words :
    delete:bool = False 
    for stopWord in stopwordsList:
        if(word.find(stopWord)!=-1):
            delete = True
            break
    if not delete:
        newWords.append(word)
words = newWords
print(len(words))
```
